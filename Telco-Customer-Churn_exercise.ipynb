```python
def encode_columns(df, columns):
    label_encoder = LabelEncoder()
    for column in columns:
        df[f"{column}_encoded"] = label_encoder.fit_transform(df[column])
    return df

# List of columns to encode
columns_to_encode = [
    "gender", "Partner", "Dependents", "PhoneService", "MultipleLines",
    "InternetService", "OnlineSecurity", "OnlineBackup", "DeviceProtection",
    "TechSupport", "StreamingTV", "StreamingMovies", "Contract",
    "PaperlessBilling", "PaymentMethod", "Churn"
]

# Encode the columns
df1 = encode_columns(df, columns_to_encode)
```

```python
df = df.drop(columns=["gender", "Partner", "Dependents", "PhoneService", "MultipleLines",
    "InternetService", "OnlineSecurity", "OnlineBackup", "DeviceProtection",
    "TechSupport", "StreamingTV", "StreamingMovies", "Contract",
    "PaperlessBilling", "PaymentMethod", "Churn"])
```

```python
df['TotalCharges'] = df['TotalCharges'].astype(float) #Este metodo no me valio y busque otra forma de cambiar de tipo de dato
```

```python
non_numeric = df[~df['TotalCharges'].apply(lambda x: x.replace('.', '', 1).isdigit())]
print(non_numeric)
```

```python
df = df.dropna(how='any') #borro todas las filas que no tengan datos
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df = df.dropna(subset=['TotalCharges'])
```

```python
df.describe()
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
#Transformo las variables numericas con Robust scaler
robust_scaler = RobustScaler()
df["tenure_scaled"] = robust_scaler.fit_transform(df[["tenure"]])
df["MonthlyCharges_scaled"] = robust_scaler.fit_transform(df[["MonthlyCharges"]])
df["TotalCharges_scaled"] = robust_scaler.fit_transform(df[["TotalCharges"]])
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.histplot(data=df, x="tenure", kde=True)
plt.title("Data sin escalar")
plt.subplot(1,2,2)
sns.histplot(data=df, x="tenure_scaled", kde=True)
plt.title("Data escalada")
plt.show()
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.histplot(data=df, x="MonthlyCharges", kde=True)
plt.title("Data sin escalar")
plt.subplot(1,2,2)
sns.histplot(data=df, x="MonthlyCharges_scaled", kde=True)
plt.title("Data escalada")
plt.show()
```

```python
#Creo una copia para probar otro scaler
df1 = df.copy()
```

```python
#Chequeo si Standard scaler es mejor
from sklearn.preprocessing import StandardScaler
standar_scaler = StandardScaler()
df1["tenure_scaled_1"] = standar_scaler.fit_transform(df1[["tenure"]])
df1["MonthlyCharges_scaled_1"] = standar_scaler.fit_transform(df1[["MonthlyCharges"]])
df1["TotalCharges_scaled_1"] = standar_scaler.fit_transform(df1[["TotalCharges"]])
```

```python
df1
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.histplot(data=df1, x="tenure", kde=True)
plt.title("Data sin escalar")
plt.subplot(1,2,2)
sns.histplot(data=df1, x="tenure_scaled_1", kde=True)
plt.title("Data normalizada")
plt.show()
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.histplot(data=df1, x="MonthlyCharges", kde=True)
plt.title("Data sin escalar")
plt.subplot(1,2,2)
sns.histplot(data=df1, x="MonthlyCharges_scaled_1", kde=True)
plt.title("Data normalizada")
plt.show()
```
